```python
from __future__ import print_function, division
import numpy as np
import pandas as pd
import pylab as pl
%pylab inline
import scipy.stats as st
```

```python
np.random.seed(0)
```

```python
n = 1000
mean = 100
stdev = np.sqrt(mean)
```

```python
pl.hist(np.random.poisson(mean, size=n), bins = 30)
pl.title('The Original randomly generated Poisson Distribution', fontsize = 16)
pl.ylabel('N', fontsize = 16)
pl.xlabel('x', fontsize = 16)
```

```python
#ploting one sample
samplesSize = 100
pl.hist(st.poisson.rvs(mu=mean, size=samplesSize))
pl.title('The Sample Poisson Distribution', fontsize = 16)
pl.ylabel('N', fontsize = 16)
pl.xlabel('x', fontsize = 16)
```

```python
samplesNumber = 100
samplesMean = np.zeros(samplesNumber)

for i in range(samplesNumber):
    samplesDist = st.poisson.rvs(mean, size=samplesSize)
    samplesMean[i] = samplesDist.mean()
    
pl.hist(samplesMean)
pl.title('The Mean of Samples Distribution', fontsize = 16)
pl.ylabel('N', fontsize = 16)
pl.xlabel('Sample Means', fontsize = 16)
```

```python
z = (mean - samplesMean)*np.sqrt(samplesNumber)/stdev
pl.hist(z)
pl.title('The Distribution of Z statistics', fontsize = 16)
pl.ylabel('N', fontsize = 16)
pl.xlabel('Z statistics', fontsize = 16)
```

```python
pl.hist(z, normed = True, label = 'Z statistics')
pl.title('The Distribution of Z statistics with fitted Gaussian Distribution', fontsize = 16)
pl.ylabel('N', fontsize = 16)
pl.xlabel('Z statistics', fontsize = 16)
fitted = st.norm.pdf(sorted(z), np.mean(sorted(z)), np.std(sorted(z)))
pl.plot(sorted(z),fitted,'-r', label = 'Fitted Gaussian Distribution')
pl.legend()
```

```python
st.kstest(z,'norm')
```

**Null hypothesis:** distribution of Z statistics of sample means generated from poisson distribution follows the Gaussian Distribution.
**Significance Level  = 0.05**.
The Kolmogorov-Smirnov statistic on given samples shows that P-value is significantly higher than significance level.
Therefore we **cannot reject the null hypothesis**.
We conclude that **Z score distribution fits the Normal Distribution**.
